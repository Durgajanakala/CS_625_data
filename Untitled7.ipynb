# **PART 1**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://data.virginia.gov/resource/vsrk-d6hx.csv")

df['Rate per 100k'] = (df['infections'] / df['population_denominator']) * 100000

df = df.drop_duplicates(subset=['week_ending_date', 'vaccination_status'], keep='first')

df['report_date'] = df['report_date'].str.slice(stop=10)

df['week_ending_date'] = df['week_ending_date'].str.slice(stop=10)

grouped = df.groupby(['week_ending_date', 'vaccination_status']).mean()

grouped = grouped.unstack()

grouped['Rate per 100k'].plot(kind='line')

plt.title('Weekly Rate of COVID-19 Cases in Virginia by Vaccination Status')
plt.xlabel('week_ending_date')
plt.ylabel('Rate per 100k')
plt.xticks(rotation=45)

plt.show()

```

# **PART 2**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://data.virginia.gov/resource/28k2-x2rj.csv")

# Filter out "Out of State" and "Not Reported" health districts
df = df[~df['health_district'].isin(['Out of State', 'Not Reported'])]

# Remove everything after Pfizer in the vaccine_manufacturer column
df['vaccine_manufacturer'] = df['vaccine_manufacturer'].str.replace('Pfizer.*', 'Pfizer')

# Filter the data to include only 1st doses
df = df[df['dose_number'] == 1]

# Group data by health district and calculate proportion of 1st doses that were Pfizer
pfizer_prop = df[df['vaccine_manufacturer'] == 'Pfizer']['dose_number'].groupby(df['health_district']).sum() / df['dose_number'].groupby(df['health_district']).sum()

# Plot the proportion of 1st doses that were Pfizer by health district
pfizer_prop.plot(kind='bar', figsize=(12,6))
plt.title("Proportion of 1st Doses that were Pfizer by Health District")
plt.xlabel("Health District")
plt.ylabel("Proportion of 1st Doses")
plt.show()
```

# **PART 3**

Find the correlation between infections from data set 1 and health region from data set .

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("https://data.virginia.gov/resource/vsrk-d6hx.csv")
infections_df = df1[['infections']]

df2 = pd.read_csv("https://raw.githubusercontent.com/Durgajanakala/public/main/New2.csv")
health_region_df = df2[['health_region']]

combined_df = pd.concat([health_region_df, infections_df], axis=1)

combined_df.plot(x='health_region', y='infections', kind='line')

plt.xlabel('Health Region')
plt.ylabel('Infections')
plt.title('Infections by Health Region')

plt.show()
```
