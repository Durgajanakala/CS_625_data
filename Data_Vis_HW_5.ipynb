# Part 1: Create Distribution Charts


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load data
df = pd.read_csv("nst-est2019-alldata.csv", encoding="ISO-8859-1")

# Filter relevant columns
df_rates = df[["NAME", "RBIRTH2019", "RDEATH2019", "RNETMIG2019"]].dropna()

# Create boxplot
sns.boxplot(data=df_rates.drop("NAME", axis=1), orient="h")
plt.xlabel("Rate")
plt.title("Distribution of Birth, Death, and Migration Rates in the US (2019)")
plt.show()

# Create eCDF and histogram for birth rates
fig = px.ecdf(df_rates, x="RBIRTH2019", color_discrete_sequence=["blue"])
fig.update_layout(title_text="Empirical CDF of Birth Rates in the US (2019)")
fig.show()

# Create the histogram
plt.hist(birth_rates, bins=15)

# Set the plot title and axis labels
plt.title('Distribution of Birth Rates in the US (2019)')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
