Question **1**: How does the overall guest satisfaction rating vary by the number of bedrooms in an Airbnb property in Vienna?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Subset the data to select only the relevant columns
satisfaction = data[['bedrooms', 'guest_satisfaction_overall']]

# Calculate the average satisfaction rating by number of bedrooms
avg_satisfaction = satisfaction.groupby('bedrooms').mean()

# Create a bar chart to visualize the average satisfaction rating by number of bedrooms
plt.figure(figsize=(12,6))
sns.barplot(x=avg_satisfaction.index, y='guest_satisfaction_overall', data=avg_satisfaction)
plt.title('Average Guest Satisfaction with Airbnb Properties in Vienna by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Overall Guest Satisfaction Rating')
plt.ylim(90, 100)
plt.show()

```

Question 2: What is the distribution of Airbnb properties in Vienna by property type and host status in the 10 smallest districts?
