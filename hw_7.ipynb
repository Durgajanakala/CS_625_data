**1.What is the average price of Airbnb listings under 300 EUR in Vienna during weekdays and weekends?**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
weekdays = pd.read_csv("vienna_weekdays.csv")
weekends = pd.read_csv("vienna_weekends.csv")

# Filter the data to prices under 300 EUR
weekdays = weekdays[weekdays['realSum'] < 300]
weekends = weekends[weekends['realSum'] < 300]

# Concatenate the datasets and create a new column indicating the day of week
data = pd.concat([weekdays.assign(day='Weekdays'), weekends.assign(day='Weekends')])

# Calculate the mean price for each day of week
means = data.groupby('day')['realSum'].mean().reset_index()

# Create a bar chart of Airbnb prices for weekdays and weekends
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='realSum', data=means, palette='Set2', width=0.5)

# Add labels and set y-axis limit
plt.title('Average Airbnb Prices in Vienna')
plt.xlabel('Day of Week')
plt.ylabel('Price (EUR)')
plt.ylim(0, 400)

# Add value labels to the bars
for i, v in enumerate(means['realSum']):
    plt.text(i, v + 10, str(round(v, 2)), ha='center')

plt.show()






```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
weekdays = pd.read_csv("vienna_weekdays.csv")
weekends = pd.read_csv("vienna_weekends.csv")

# Filter the data to prices under 300 EUR
weekdays = weekdays[weekdays['realSum'] < 300]
weekends = weekends[weekends['realSum'] < 300]

# Create a box plot of Airbnb prices for weekdays and weekends
plt.figure(figsize=(8, 6))
sns.boxplot(x='weekday', y='realSum', data=pd.concat([weekdays.assign(weekday='Weekdays'), weekends.assign(weekday='Weekends')]), hue='weekday', palette='Set2')
plt.title('Distribution of Airbnb Prices in Vienna')
plt.xlabel('Day of Week')
plt.ylabel('Price (EUR)')
plt.ylim(0, 400)
plt.show()
```

**What is the average nightly price of each room type (shared, private) for superhost and non-superhost hosts on weekdays and weekends in Vienna?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
weekdays = pd.read_csv("vienna_weekdays.csv")
weekends = pd.read_csv("vienna_weekends.csv")

# Group the data by room type and host is superhost and count the number of occurrences
weekdays_by_type_status = weekdays.groupby(['room_type', 'host_is_superhost'])['room_type'].count().reset_index(name='count')
weekends_by_type_status = weekends.groupby(['room_type', 'host_is_superhost'])['room_type'].count().reset_index(name='count')

# Set the color palette and font sizes
sns.set_palette("pastel")
sns.set(font_scale=1.2)

# Create a line chart showing the number of occurrences of each room type and host is superhost on weekdays and weekends
fig, axs = plt.subplots(ncols=2, figsize=(12, 6))
sns.lineplot(x="room_type", y="count", hue="host_is_superhost", data=weekdays_by_type_status, ax=axs[0])
axs[0].set_title('Weekday Room Type and Host is Superhost Count', fontsize=14)
axs[0].set_xlabel('Room Type', fontsize=12)
axs[0].set_ylabel('Count', fontsize=12)
axs[0].legend(title='Superhost', fontsize=10, title_fontsize=12)
sns.lineplot(x="room_type", y="count", hue="host_is_superhost", data=weekends_by_type_status, ax=axs[1])
axs[1].set_title('Weekend Room Type and Host is Superhost Count', fontsize=14)
axs[1].set_xlabel('Room Type', fontsize=12)
axs[1].set_ylabel('Count', fontsize=12)
axs[1].legend(title='Superhost', fontsize=10, title_fontsize=12)
fig.suptitle("Number of Occurrences of Each Room Type and Host is Superhost on Weekdays and Weekends", fontsize=16)
plt.tight_layout()
plt.show()


```
