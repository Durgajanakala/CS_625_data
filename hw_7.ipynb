**1.What is the average price of Airbnb listings under 300 EUR in Vienna during weekdays and weekends?**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
weekdays = pd.read_csv("vienna_weekdays.csv")
weekends = pd.read_csv("vienna_weekends.csv")

# Filter the data to prices under 300 EUR
weekdays = weekdays[weekdays['realSum'] < 300]
weekends = weekends[weekends['realSum'] < 300]

# Create a box plot of Airbnb prices for weekdays and weekends
plt.figure(figsize=(8, 6))
sns.boxplot(x='weekday', y='realSum', data=pd.concat([weekdays.assign(weekday='Weekdays'), weekends.assign(weekday='Weekends')]), hue='weekday', palette='Set2')
plt.title('Distribution of Airbnb Prices in Vienna')
plt.xlabel('Day of Week')
plt.ylabel('Price (EUR)')
plt.ylim(0, 400)
plt.show()
```

**2.What is the distribution of Airbnb properties in Vienna by property type and host status in the 10 smallest districts in weekdays and weekends in Vienna?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
weekdays = pd.read_csv("vienna_weekdays.csv")
weekends = pd.read_csv("vienna_weekends.csv")

# Group the data by room type and count the number of occurrences
weekdays_by_room_type = weekdays.groupby('room_type')['room_type'].count().reset_index(name='count')
weekends_by_room_type = weekends.groupby('room_type')['room_type'].count().reset_index(name='count')

# Create a pie chart showing the percentage of each room type on weekdays and weekends
fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
axs[0].pie(weekdays_by_room_type['count'], labels=weekdays_by_room_type['room_type'], autopct='%1.1f%%')
axs[0].set_title('Weekday Room Type Percentage')
axs[1].pie(weekends_by_room_type['count'], labels=weekends_by_room_type['room_type'], autopct='%1.1f%%')
axs[1].set_title('Weekend Room Type Percentage')
fig.suptitle("Percentage of Each Room Type on Weekdays and Weekends")
plt.show()
```
